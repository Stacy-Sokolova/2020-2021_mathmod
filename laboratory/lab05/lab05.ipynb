```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
a = 0.22
b = 0.044
c = 0.33
d = 0.022
```

```python
def syst(x, t):
    return np.array([-a*x[0]+b*x[0]*x[1], c*x[1]-d*x[0]*x[1]])
```

```python
t = np.linspace(0, 400, 4000)
x0 = [3, 8]

y = odeint(syst, x0, t)
```

```python
plt.plot(t, y[:, 0])
plt.plot(t, y[:, 1])
plt.title("       ")
plt.xlabel("t")
plt.ylabel(" ")
plt.legend(["",""])
plt.grid()
plt.show()
```

```python
plt.plot(y[:, 0], y[:, 1])
plt.scatter(c/d, a/b, color='orange') # 
plt.title("        ")
plt.xlabel("")
plt.ylabel("")
plt.grid()
plt.show()
```
